# Place chip designs

```python
Grid_sweep_type = "gridsweep"
Grid_sweep_frequency = [[6500, 7500], [7000, 8000], [7500, 8500], [8000, 9000]]
Grid_sweep_dummy     = [0, 0, 0, 0]
# Grid_sweep_type = "array"
# Grid_sweep_array = [
#     [6500, 7500], None        , [7500, 8500], None        ,
#     [7000, 8000], None        , [8000, 9000], None        ,
#     None        , [6500, 7500], None        , [7500, 8500],
#     None        , [7000, 8000], None        , [8000, 9000]
# ]
Grid_gap_x = 1
Grid_gap_y = 1

if Grid_sweep_type == "array":
    device_list = []
    for row in Grid_sweep_array:
        for cell in row:
            if cell is None:
                device_list.append(FM)
            else:
                device_list.append( custom_chip(None, cell))
    D = pg.grid(
        device_list,
        spacing = (Grid_gap_x * config["Frame_size_width"], Grid_gap_y * config["Frame_size_height"]),
        shape = np.array(Grid_sweep_array, dtype=object).shape
    )
elif Grid_sweep_type == "gridsweep":
    D = pg.gridsweep(
        function = custom_chip,
        param_x = {'width' : Grid_sweep_dummy},
        param_y = {'height' : Grid_sweep_frequency},
        spacing = (Grid_gap_x * config["Frame_size_width"], Grid_gap_y * config["Frame_size_height"]),
        label_layer = None
        )
    ## Remove corners
    for iref, ref in enumerate(D.references[:]):
        if iref == 0 or iref == len(Grid_sweep_dummy)-1 or iref == len(Grid_sweep_dummy) * (len(Grid_sweep_frequency)-1) or iref == len(Grid_sweep_dummy) * len(Grid_sweep_frequency) - 1:
            D.remove(ref)

D.center = (0,0)

wafer.add_ref(D)
qp(D)
```
